# Encoding : 
It is the process to convert categorical(object) data into numerical data or inverse of this process to better understanding of data

Encoding is in the form of mathematical data

Nominal data : in this data is not connected or not in sequential form

Ordinal data : which is in the sequential order or connected 

```python
import pandas as pd
dataset = pd.read_csv("loan.csv")
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
dataset['Married'].fillna(dataset['Married'].mode()[0], inplace=True)
```

```python
en_data = dataset[['Gender','Married']]
en_data
```

```python
pd.get_dummies(en_data).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
ar = ohe.fit_transform(en_data).toarray()
#ohe.fit_transform(en_data): output : -sparse matrix is the matrix which contain only 0,1 value
```

```python
pd.DataFrame(ar,columns=['Gender_Female','Gender_Male','Married_No', 'Married_Yes'])
```

```python
ohe = OneHotEncoder(drop="first")
ar = ohe.fit_transform(en_data).toarray()
#ohe.fit_transform(en_data): output : -sparse matrix is the matrix which contain only 0,1 value
```

```python
pd.DataFrame(ar,columns=['Gender_Male', 'Married_Yes'])
```

# Ordinal Encoding

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
df.head(5)
                  
```

```python
ord_data = [["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories = ord_data)
oe.fit(df[["Size"]])
```

```python
df["Size_en"] = oe.transform(df[["Size"]])
```

```python
df
```

# Map method

```python
ord_data2 = {"s":0,"m":1,"l":2,"xl":3} #we can decide the transform value in map
```

```python
df["Size_en_map"] = df["Size"].map(ord_data2)
```

```python
df
```
