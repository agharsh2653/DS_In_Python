```python
bike_brand = list(df['brand'].value_counts().head(16).index)
bike_brand
```

```python
df2 = df[df['brand'].isin(bike_brand)]
df2.head()
```

```python
list(df2['brand'].value_counts().index)
```

```python
dt = {'Bajaj':1,
 'Royal Enfield':2,
 'Hero':3,
 'Honda':4,
 'Yamaha':5,
 'TVS':6,
 'KTM':7,
 'Suzuki':8,
 'Harley-Davidson':9,
 'Kawasaki':10,
 'Hyosung':11,
 'Mahindra':12,
 'Benelli':13,
 'Triumph':14,
 'Ducati':15,
 'BMW':16}
```

```python
df2['brand'] = df['brand'].map(dt)
df2
```

```python
df2.drop(['bike_name','city'],axis=1,inplace=True)
df2
```

```python
df2['brand'].value_counts()
```

```python
x = df2.drop('price',axis='columns')
y = df2[['price']]

x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)  # test size 20%
```

```python
# Importing linearregression algorithm class
from sklearn.linear_model import LinearRegression
```

```python
# Instance creation

lnr = LinearRegression()
```

```python
## Algorithm train

lnr.fit(x_train,y_train)
```

```python
x_train.shape
```

```python
x_test
```

```python
lnr.score(x_train,y_train)*100
```

```python
lnr.score(x_test,y_test)*100
```

```python
prediction=lnr.predict(x_test)
```
