Melt()

```python
import pandas as pd
```

```python
var = pd.DataFrame({"days":[1,2,3,4,5,6],"english":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})
var
```

```python
pd.melt(var)
```

```python
#pd.melt(var,id_vars=["days"])
pd.melt(var,id_vars=["days"], var_name="subjects",value_name="marks")
```

Pivot()

```python
var1 = pd.DataFrame({"days":[1,2,3,4,5,6],
                     "st_name":["a","b","c","a","b","c"],
                     "english":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})
var1
```

```python
var1.pivot(index="days",columns="st_name",values="maths")
```

```python
var2 = pd.DataFrame({"days":[1,2,1,1,1,2],
                     "st_name":["a","b","b","a","b","a"],
                     "english":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})
var2
```

```python
var2.pivot_table(index="days",columns="st_name",values="maths",aggfunc="mean")
```

```python
var2.pivot_table(index="days",columns="st_name",aggfunc="mean")
```

```python
var2.pivot_table(index="days",columns="st_name",aggfunc="mean",margins="True")
```
