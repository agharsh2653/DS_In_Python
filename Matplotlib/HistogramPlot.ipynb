# Histogram Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import random 

x = np.random.randint(10,60,(50))
print(x)
```

```python
no = [16, 44, 46, 42, 34, 12, 39, 41, 48, 34, 42, 35, 40, 35, 28, 22, 36, 15, 23, 26, 29, 52, 37,
 35, 31, 34, 22, 48, 46, 55, 45, 37, 13, 57, 15, 29, 24, 41, 36, 40, 15, 12, 18 ,40, 14, 21, 46,
 52, 43]
plt.hist(no,"auto",(0,100),color="gold",edgecolor="r",linewidth=2,bottom=0)
#plt.hist(no,"auto",(0,100),color="gold",edgecolor="r",linewidth=2,cumulative=-1) in this cumulative parameter print the graph in reverse order
# plt.hist(no,bins=10,color="gold",edgecolor="r",linewidth=2) bins=10 is the parameter which show the data visualization in no of visualization
plt.title("Histogram Plot")
plt.xlabel("Python")
plt.ylabel("Value")
plt.show()
```
