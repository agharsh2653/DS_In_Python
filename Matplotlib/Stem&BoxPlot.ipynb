# Stem Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,5,6,3,3]
plt.stem(x,y,linefmt=":",markerfmt="ro",basefmt="gold",bottom=0,label="Python",orientation='horizontal')
plt.legend()
plt.show()
```

```python

# importing libraries 
import matplotlib.pyplot as plt 
import numpy as np 
  
x = np.linspace(0.1, 2 * np.pi, 41) 
y = np.exp(np.sin(x)) 
  
plt.stem(x, y) 
plt.show()
```

# BOX PLOT

whisky plot is the created line on the upper side of box 


```python
b = [10,20,30,40,50,60,70] 
# plt.boxplot(b,notch=True,vert=True,widths=0.4,tick_labels=["python"],patch_artist=True)
plt.boxplot(b,widths=0.2,tick_labels=['python'],showmeans=True)
plt.show()
```

dot it the outler : It is created when the point is instantly far in the below graph we increase index of b 120 

```python
b = [10,20,30,40,50,60,70,120 ]
# plt.boxplot(b,notch=True,vert=True,widths=0.4,tick_labels=["python"],patch_artist=True)
plt.boxplot(b,widths=0.2,tick_labels=['python'],showmeans=True,sym="g+",boxprops=dict(color="gold"),capprops=dict(color="r"),
           whiskerprops=dict(color="b"),flierprops=dict(markeredgecolor="r"))
plt.show()
```

To connect the wisky plot we use whis parameter

```python
b = [10,20,30,40,50,60,70,120 ]
# plt.boxplot(b,notch=True,vert=True,widths=0.4,tick_labels=["python"],patch_artist=True)
plt.boxplot(b,widths=0.2,tick_labels=['python'],showmeans=True,whis=2)
plt.show()
```

```python
b = [10,20,30,40,50,60,70]
a = [10,20,40,50,30,60,90]
y=[a,b]

plt.boxplot(y,widths=0.2,tick_labels=['python','java'],showmeans=True,sym="g+",boxprops=dict(color="gold"),capprops=dict(color="r"),
           whiskerprops=dict(color="b"),flierprops=dict(markeredgecolor="r"))
plt.show()
```
