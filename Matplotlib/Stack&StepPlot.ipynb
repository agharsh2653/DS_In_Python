# Stack Plot

```python
import matplotlib.pyplot as  plt

x= [1,2,3,4,5]
area1 = [2,4,2,8,6]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]

l = ['area1','area2','area3']
plt.stackplot(x,area1, area2,area3,labels=l,colors=["r","g","gold"],baseline="zero")

plt.title("python")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as  plt

x= [1,2,3,4,5]
area1 = [2,4,2,8,6]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]

l = ['area1','area2','area3']
plt.stackplot(x,area1, area2,area3,labels=l,colors=["r","g","gold"],baseline="wiggle")

plt.title("python")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

# Step Plot

```python
import matplotlib.pyplot as plt                                                          
x= [1,2,3,4,5]
y =[2,2,4,5,3]

plt.step(x,y,color="r",marker="o",ms=10,mfc="gold",label="python")
plt.title("python")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid()
plt.legend(loc=2)
plt.show()
```
