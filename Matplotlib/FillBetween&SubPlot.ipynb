# Fill_Between Plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [0,2,1,4,1]
plt.plot(x,y)
plt.fill_between(x=[2,4],y1=2,y2=4,color="gold")

plt.title("Python")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

 where does not work on list it work on int on the parameter 'where=(x>=2) &(x<=4)'    
 ERROR :- '>=' not supported between instances of 'list' and 'int'

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([0,2,1,4,1])
plt.plot(x,y)
plt.fill_between(x,y,color="gold",where=(x>=2) & (x<=4),alpha=0.5)

plt.title("Python")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

# Sub Plot

```python
x= [1,2,3,4]
y= [1,2,3,4]
plt.subplot(2,2,1)
plt.plot(x,y,color="r")

plt.subplot(2,2,2)
plt.pie([1],colors="g")

plt.subplot(2,2,3)
plt.pie([10,20,30,40])

x2 = ['a','s','d','h']
y2 = [10,20,30,40]
plt.subplot(2,2,4)
plt.bar(x2,y2)

plt.show()
```
