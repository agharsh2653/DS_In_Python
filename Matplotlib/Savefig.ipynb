# Savefig 

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([0,2,1,4,1])
plt.plot(x,y)
plt.fill_between(x,y,color="gold",where=(x>=2) & (x<=4))

plt.title("Python")
plt.xlabel("x axis")
plt.ylabel("y axis")

#plt.savefig("sub")
#plt.savefig("sub1",dpi=200,facecolor="g")
#dpi is depth per inch increase the quality of images like pixels
#facecolor fill the outside part of x and y axis
# plt.savefig("sub12.pdf",dpi=200,facecolor="g")
# plt.savefig("sub123.jpg",dpi=200,facecolor="g",transparent=False)
plt.savefig("sub2.jpg",dpi=200,facecolor="g",transparent=True)
# transparent remove the white color and fill the facecolor in the graph if include
plt.savefig("sub3.jpg",dpi=200,facecolor="g",transparent=False,bbox_inches="tight")

plt.show()
```
