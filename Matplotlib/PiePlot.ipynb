# Pie Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=["C","Python","C++","Java"]
ex=[0.2,0.0,0.1,0.0]

c = ["royalblue","gold","navy","teal"]

plt.pie(x,labels=y,colors=c, explode=ex,autopct="%0.1f%%",shadow=True,radius=0.7,labeldistance=1.2)
plt.legend()
plt.show()
```

```python
x=[10,20,30,40]
y=["C","Python","C++","Java"]
ex=[0.2,0.0,0.1,0.0]

c = ["royalblue","gold","navy","teal"]

plt.pie(x,labels=y,colors=c, explode=ex,autopct="%0.1f%%",shadow=True,radius=0.7,labeldistance=1.2,startangle=90
       ,textprops={"fontsize":15})
plt.legend(loc=4)
plt.show()
```

```python
x=[10,20,30,40]
y=["C","Python","C++","Java"]
ex=[0.2,0.0,0.1,0.0]

c = ["royalblue","gold","navy","teal"]

plt.pie(x,labels=y,colors=c, explode=ex,autopct="%0.1f%%",shadow=True,radius=1.1,labeldistance=1.2,startangle=0
       ,textprops={"fontsize":15},counterclock=False, wedgeprops={"linewidth":4, "edgecolor":"r"},
       center=(1,2))

#linewidth increase the shadow and counterclock set false the it rotate it width=2 it increase color size inside it overlap each other

plt.show()
```

```python
x=[10,20,30,40]
y=["C","Python","C++","Java"]
ex=[0.2,0.0,0.1,0.0]

c = ["royalblue","gold","navy","teal"]

plt.pie(x,labels=y,colors=c, explode=ex,autopct="%0.1f%%",shadow=True,radius=1.1,labeldistance=1,startangle=0
       ,textprops={"fontsize":15},counterclock=False, wedgeprops={"linewidth":4, "edgecolor":"r"},
       rotatelabels=True)

#linewidth increase the shadow and counterclock set false the it rotate it width=2 it increase color size inside it overlap each other
plt.title("Pie Plot")
plt.show()
```

```python
plt.pie([1])
plt.show()
```

```python
x=[10,20,30,40]
x1= [40,30,20,10]
y=["C","Python","C++","Java"]

c = ["royalblue","gold","navy","teal"]

plt.pie(x,labels=y,colors=c,radius=1.5)
plt.pie([1],colors="w")

plt.show()
```

```python
x=[10,20,30,40]
x1= [40,30,20,10]
y=["C","Python","C++","Java"]

c = ["r","b","g","y"]

plt.pie(x,labels=y,colors=c,radius=1.5)
cr=plt.Circle(xy=(0,0),radius=1,facecolor="w")
plt.gca().add_artist(cr)

plt.show()
```
