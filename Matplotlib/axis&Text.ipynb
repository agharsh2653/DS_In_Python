# xticks(),yticks(),xlim(),ylim(),axis()

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

plt.xticks(x,labels=["python","c","java","c++","js"])
plt.yticks(x)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)


plt.axis([0,10,0,7])
plt.show()
```

# Wroking with text in Matplotlib

```python
import matplotlib.pyplot as plt
language = [1,2,3,4,5]
popular = [3,1,5,2,6]

plt.plot(language,popular)
plt.xlabel("Languages",fontsize=15)
plt.ylabel("Popular Score",fontsize=10)
plt.title("Wscube",fontsize=15)

plt.text(2,5,"Java",fontsize=15,style="italic",bbox={"facecolor":"navy"})

plt.annotate("python",xy=(1,3),xytext=(2,4),arrowprops=dict(facecolor="black",shrink=100))

plt.legend(["up"],loc=9,facecolor="gold",edgecolor="r",framealpha=0.7,shadow=True)

plt.show()
```
