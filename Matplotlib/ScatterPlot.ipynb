## Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
day= [1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
plt.scatter(day,no,color="r")
plt.title("Scatter Plot",color="teal" ,fontsize=15)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

```python
day= [1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors= ["r","g","b","y","b","teal","gold"]
size= [400,300,200,600,500,200,250]
plt.scatter(day,no,c=colors,s=size,alpha=0.9,marker="*",edgecolor="gold",linewidth=2)
plt.title("Scatter Plot",color="teal" ,fontsize=15)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

```python
day= [1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
no2=[3,4,1,5,2,7,9]
#colors= ["r","g","b","y","b","teal","gold"]
colors= [29,78,34,87,56,72,45]
size= [400,300,200,600,500,200,250]
plt.scatter(day,no,c=colors,s=size,cmap="viridis",alpha=0.7)
plt.scatter(day,no2,c=colors,s=size,cmap="copper",alpha=0.7)
#cmap name BrBG, copper, Assent,viridis
t=plt.colorbar()
t.set_label("ColorBar", fontsize=15)
plt.title("Scatter Plot",color="teal" ,fontsize=15)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No",fontsize=15)


plt.show()
```
